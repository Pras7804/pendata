# **K-Means Clustering**

## **Apa itu K-Means Clustering?**
**K-Means clustering** adalah algoritma pembelajaran tanpa pengawasan yang digunakan untuk pengelompokan data, yang mengelompokkan titik data yang tidak berlabel ke dalam kelompok atau kluster.

**K-means** adalah algoritma pengelompokan berbasis centroid berulang yang mempartisi kumpulan data menjadi kelompok serupa berdasarkan jarak antara centroid mereka.


Meskipun ada berbagai jenis algoritma pengelompokan, termasuk *eksklusif, tumpang tindih, hirarkis, dan probabilistik*, algoritma k-means clustering adalah contoh metode pengelompokan eksklusif atau "keras". Bentuk pengelompokan ini menetapkan bahwa titik data dapat ada hanya dalam satu kluster. Jenis analisis kluster ini umumnya digunakan dalam ilmu data untuk segmentasi pasar, pengelompokan dokumen, segmentasi gambar, dan kompresi gambar.

## **Cara kerja K-Means Clustering**
1.   Menentukan jumlah kluster **K**
2.   Pilih **K centroid awal** secara acak.
3.   Hitung jarak setiap data ke tiap centroid.
4.   Kelompokkkan data berdasarkan centroid terdekat
5.   Hitung ulang centroid dari rata-rata anggota kluster.
6.   Ulangi langkah 3-5 hingga konvergen (tidak ada perubahan signifikan) atau jumlah iterasi maksimum tercapai


## **Keuntungan dan Kekurangan**
**Keuntungan**
1.   Cepat dan efisien
2.   Mudah dipahami dan diimplementasikan
3.   Berkerja baik pada data besar
4.   Hasil stabil jika inisialisasi bagus


**Kekurangan**
1.   Harus menentukan nilai **K** di awal
2.   Sensitif terhadap outlier
3.   Dipengaruhi oleh skala dan dimensi data
4.   Hanya cocok untuk bentuk kluster sferis



## **Evaluasi Hasil Clustering**
*   **Inertia:** Jumlah kuadrat jarak antara titik dan centroid.
$$
\text{Inertia} = \sum_{i=1}^{n} \left\| x_i - \mu_{c_i} \right\|^2
$$
dimana:
    - $x_i$: TItik data ke-*i*
    - $\mu_{c_i}$: Centroid dari cluster tempat $x_i$ berada

*   **Silhouette Score:** Ukuran seberapa mirip suatu objek dengan klusternya dibandding kluster lain.
$$
s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}
$$
Keterangan:
    - $a(i)$: Rata-rata jarak titik *i* ke semua titik lain dalam **klaster yang sama**
    - $b(i)$: Rata-rata jarak titik *i* ke semua titik dalam **klaster terdekat (tetangga)**


## **Contoh Kasus**

| Titik | X | Y |
|----------------------------------------|--------------------------------|-------------|
| A                                     | 1                        | 2      |
| B                                      | 1                       | 4      |
| C                                     | 5                        | 2      |
| D                                      | 5                        | 4      |


**Tujuan:** Kelompokkan data menjadi **k = 2** cluster

### Langkah 1: Inisialisasi centroid secara acak

Misalnya kita pilih:
*   Centroid 1 (C1) = titik A --> (1,2)
*   Cemtroid 2 (C2) = titik C --> (5,2)

### Langkah 2: Hitung jarak setiap titik ke **C1** dan **C2**
Menggunakan rumus Euclidean distance:

$$
d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}
$$

**Titik A(1,2)**
*   ke $
C1 = \sqrt{(1 - 1)^2 + (2 - 2)^2}=0
$
*   ke $
C2 = \sqrt{(5 - 1)^2 + (2 - 2)^2} \sqrt{16} = 4
$

    --> Masuk ke **Cluster 1**


**Titik B(1,4)**
*   ke $
C1 = \sqrt{(1 - 1)^2 + (4 - 2)^2} = \sqrt{4} = 2
$
*   ke $
C2 = \sqrt{(5 - 1)^2 + (2 - 4)^2} = \sqrt{20} \approx 4.47
$
   
    --> Masuk ke **Cluster 1**



**Titik C(5,2)**
*   Ke C1 = 4 (sudah dihitung)
*   Ke C2 = 0
    
    --> Masuk ke **Cluster 2**



**Titik D(5,4)**
*   Ke $
C1 = \sqrt{(5 - 1)^2 + (4 - 2)^2} = \sqrt{20} \approx 4.47
$
*   $
C2 = \sqrt{(5 - 5)^2 + (4 - 2)^2} = \sqrt{4} = 2
$
    
    --> Masuk ke **Cluster 2**


**Hasil Cluster Awal:**
*   **Cluster 1:** A(1,2), B(1,4)
*   **Cluster 2:** C(5,2), D(5,4)



### Langkah 3: Update centroid
Cluster 1 (A, B)
*   Centroid = rata-rata:
    
    X = ( 1 + 1 ) / 2 = 1, Y = ( 2 + 4 ) / 2 = 3

    --> **New C1** = (1,3)


Cluster 2 (C, D)
*   Centroid:

    X = ( 5 + 5 ) / 2 = 5, Y = ( 2 + 4 ) / 2 = 3

    --> **New C2** = (5,3)



### Langkah 4: Hitung ulang jarak ke centroid baru

Lakukan lagi langkah perhitungan jarak untuk semua titik ke centroid baru:

Misalnya titik A (1,2) ke C1 (1,3) = $ \sqrt{1} = 1
$
â€¦ dan seterusnya.

**Tapi:** Karena hasil klasternya tidak berubah, maka algoritma konvergen, dan proses berhenti.

### Hasil Akhir:


*   **Cluster 1:** A(1,2), B(1,4)
*   **Cluster 2:** C(5,2), D(5,4)

### Membuat fungsi bantu inisialisasi kmeans

**Membuat fungsi untuk inisialisasi kmeans**

Dengan paramater kunci:
*   *n_clusters=K*: jumlah cluster yang ingin dicoba.
*   *init='k-means++'*: cara inisialisasi centroid agar lebih stabil.
*   *n_init*: berapa kali menjalankan algoritma dengan inisialisasi berbeda.
*   *max_iter, tol*: batas iterasi dan toleransi konvergensi.
*   *tol=0.0001*: Toleransi untuk konvergensi. Jika perubahan dalam objective function kurang dari nilai ini, pelatihan berhenti.
*   *verbose=0*: Tidak mencetak log ke terminal.
*   *random_state=42*: memastikan hasilnya konsisten setiap kali dijalankan.


```python
def initialize_kmeans(K):
    return KMeans(
        n_clusters=K,
        init='k-means++',
        n_init='auto',
        max_iter=700,
        tol=0.0001,
        verbose=0,
        random_state=42,
        copy_x=True,
        algorithm='lloyd'
    )
```

### Perbandingan

Melakukan perbandingan clustering dan evaluasi untuk tiap K (2, 3, 4)

#### K-Means untuk K = 2

Inisialisasi & Fit

Dengan inisialisasi K = 2

```python
kmeans_2 = initialize_kmeans(2)
labels_2 = kmeans_2.fit_predict(X)
```

Tampilkan Label Cluster

```python
df_k2 = df.copy()
df_k2['Cluster_2'] = labels_2
df_k2 = df_k2.drop(columns=['Class'])
df_k2.head(150)
```

Tampilkan Centroid Akhir (Skala Asli)

```python
centroid_df2 = pd.DataFrame(
    kmeans_2.cluster_centers_,
    columns=feature_cols
)
centroid_df2.index.name = 'Cluster_2'
print("Centroid akhir (skala asli) untuk K=2:")
display(centroid_df2)
```

Evaluasi

```python
print(f"Inertia          = {kmeans_2.inertia_:.2f}")
print(f"Silhouette Score = {silhouette_score(X, labels_2):.4f}")
```

### Visualisasi

**Mengapa menggunakan PCA?**
karena agar bisa Mereduksi 4 dimensi Iris menjadi 2 dimensi, agar bisa dipetakan di plot 2D, sambil mempertahankan sebanyak mungkin varian data.


```python
from sklearn.decomposition import PCA

# 1. Reduksi dimensi ke 2 komponen dengan PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

# 2. Proyeksikan centroid (skala standarisasi) juga ke ruang PCA
centroids_2_pca = pca.transform(kmeans_2.cluster_centers_)
centroids_3_pca = pca.transform(kmeans_3.cluster_centers_)
centroids_4_pca = pca.transform(kmeans_4.cluster_centers_)
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_2, alpha=0.6, edgecolor='k', label='Data')
plt.scatter(centroids_2_pca[:,0], centroids_2_pca[:,1],
            marker='X', s=200, linewidths=2, edgecolor='k', label='Centroid')
plt.title('PCA Scatter Plot â€” K = 2')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_3, alpha=0.6, edgecolor='k', label='Data')
plt.scatter(centroids_3_pca[:,0], centroids_3_pca[:,1],
            marker='X', s=200, linewidths=2, edgecolor='k', label='Centroid')
plt.title('PCA Scatter Plot â€” K = 3')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.grid(True)
plt.show()
```

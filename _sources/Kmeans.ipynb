# **K-Means Clustering**

## **Apa itu K-Means Clustering?**
**K-Means clustering** adalah algoritma pembelajaran tanpa pengawasan yang digunakan untuk pengelompokan data, yang mengelompokkan titik data yang tidak berlabel ke dalam kelompok atau kluster.

**K-means** adalah algoritma pengelompokan berbasis centroid berulang yang mempartisi kumpulan data menjadi kelompok serupa berdasarkan jarak antara centroid mereka.


Meskipun ada berbagai jenis algoritma pengelompokan, termasuk *eksklusif, tumpang tindih, hirarkis, dan probabilistik*, algoritma k-means clustering adalah contoh metode pengelompokan eksklusif atau "keras". Bentuk pengelompokan ini menetapkan bahwa titik data dapat ada hanya dalam satu kluster. Jenis analisis kluster ini umumnya digunakan dalam ilmu data untuk segmentasi pasar, pengelompokan dokumen, segmentasi gambar, dan kompresi gambar.

## **Cara kerja K-Means Clustering**
1.   Menentukan jumlah kluster **K**
2.   Pilih **K centroid awal** secara acak.
3.   Hitung jarak setiap data ke tiap centroid.
4.   Kelompokkkan data berdasarkan centroid terdekat
5.   Hitung ulang centroid dari rata-rata anggota kluster.
6.   Ulangi langkah 3-5 hingga konvergen (tidak ada perubahan signifikan) atau jumlah iterasi maksimum tercapai


## **Keuntungan dan Kekurangan**
**Keuntungan**
1.   Cepat dan efisien
2.   Mudah dipahami dan diimplementasikan
3.   Berkerja baik pada data besar
4.   Hasil stabil jika inisialisasi bagus


**Kekurangan**
1.   Harus menentukan nilai **K** di awal
2.   Sensitif terhadap outlier
3.   Dipengaruhi oleh skala dan dimensi data
4.   Hanya cocok untuk bentuk kluster sferis



## **Evaluasi Hasil Clustering**
*   **Inertia:** Jumlah kuadrat jarak antara titik dan centroid.

    $$
    \text{Inertia} = \sum_{i=1}^{n} \left\| x_i - \mu_{c_i} \right\|^2
    $$

      dimana:
    
      - $x_i$: TItik data ke-*i*
      - $\mu_{c_i}$: Centroid dari cluster tempat $x_i$ berada

*   **Silhouette Score:** Ukuran seberapa mirip suatu objek dengan klusternya dibandding kluster lain.

    $$
    s(i) = \frac{b(i) - a(i)}{\max\{a(i), b(i)\}}
    $$

      Keterangan:
    
      - $a(i)$: Rata-rata jarak titik *i* ke semua titik lain dalam **klaster yang sama**
      - $b(i)$: Rata-rata jarak titik *i* ke semua titik dalam **klaster terdekat (tetangga)**


## **Contoh Kasus**

| Titik | X | Y |
|----------------------------------------|--------------------------------|-------------|
| A                                     | 1                        | 2      |
| B                                      | 1                       | 4      |
| C                                     | 5                        | 2      |
| D                                      | 5                        | 4      |


**Tujuan:** Kelompokkan data menjadi **k = 2** cluster

### Langkah 1: Inisialisasi centroid secara acak

Misalnya kita pilih:
*   Centroid 1 (C1) = titik A --> (1,2)
*   Cemtroid 2 (C2) = titik C --> (5,2)

### Langkah 2: Hitung jarak setiap titik ke **C1** dan **C2**
Menggunakan rumus Euclidean distance:

$$
d = \sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}
$$

**Titik A(1,2)**
*   ke $
C1 = \sqrt{(1 - 1)^2 + (2 - 2)^2}=0
$
*   ke $
C2 = \sqrt{(5 - 1)^2 + (2 - 2)^2} \sqrt{16} = 4
$

    --> Masuk ke **Cluster 1**


**Titik B(1,4)**
*   ke $
C1 = \sqrt{(1 - 1)^2 + (4 - 2)^2} = \sqrt{4} = 2
$
*   ke $
C2 = \sqrt{(5 - 1)^2 + (2 - 4)^2} = \sqrt{20} \approx 4.47
$
   
    --> Masuk ke **Cluster 1**



**Titik C(5,2)**
*   Ke C1 = 4 (sudah dihitung)
*   Ke C2 = 0
    
    --> Masuk ke **Cluster 2**



**Titik D(5,4)**
*   Ke $
C1 = \sqrt{(5 - 1)^2 + (4 - 2)^2} = \sqrt{20} \approx 4.47
$
*   $
C2 = \sqrt{(5 - 5)^2 + (4 - 2)^2} = \sqrt{4} = 2
$
    
    --> Masuk ke **Cluster 2**


**Hasil Cluster Awal:**
*   **Cluster 1:** A(1,2), B(1,4)
*   **Cluster 2:** C(5,2), D(5,4)



### Langkah 3: Update centroid
Cluster 1 (A, B)
*   Centroid = rata-rata:
    
    X = ( 1 + 1 ) / 2 = 1, Y = ( 2 + 4 ) / 2 = 3

    --> **New C1** = (1,3)


Cluster 2 (C, D)
*   Centroid:

    X = ( 5 + 5 ) / 2 = 5, Y = ( 2 + 4 ) / 2 = 3

    --> **New C2** = (5,3)



### Langkah 4: Hitung ulang jarak ke centroid baru

Lakukan lagi langkah perhitungan jarak untuk semua titik ke centroid baru:

Misalnya titik A (1,2) ke C1 (1,3) = $ \sqrt{1} = 1
$
â€¦ dan seterusnya.

**Tapi:** Karena hasil klasternya tidak berubah, maka algoritma konvergen, dan proses berhenti.

### Hasil Akhir:


*   **Cluster 1:** A(1,2), B(1,4)
*   **Cluster 2:** C(5,2), D(5,4)

### **2. Preprocessing Data**

#### **Normalisasi**

##### **Mengapa Normalisasi Diperlukan?**

Masalah utamanya ada pada fitur-fitur dari dataset iris seperti; `sepal_length`, `sepal_width`, `petal_legth`, `petal_width` memiliki rentang nilai yang berbeda-beda, Misalnya:

*   `sepal_length` bisa berkisar antara 4.3 sampai 7.9
*   `petal_width` hanya antara 0.1 sampai 2.5

Jika tidak dinormalisasi, fitur yang memiliki **skala lebih besar** akan **mendominasi** perhitungan jarak (seperti *Euclidean Distance*) dalam K-means.

Maka dari itu saya menormalisasikannya menggunakan Min-Max Scaler untuk mengubah setiap fitur ke rentang 0 hingga 1 yang berfungsi untuk:

*   Menghilangkan dominasi fitur besar
*   Memastikan semua fitur berkontribusi secara adil dalam proses clustering.




````python
# Fitur yang akan digunakan (kecuali id dan class)
features = df_combined.drop(columns=['id', 'class'])

# Inisialisasi dan terapkan MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(features)

# Konversi ke DataFrame untuk kejelasan
df_scaled = pd.DataFrame(X_scaled, columns=features.columns)

print("Data setelah normalisasi:")
print(df_scaled.head())
````

#### **Reduksi**

##### **Mengapa Perlu Redukasi Dimensi (PCA)?**

Masalah utamanya ada pada dataset iris yang memiliki **4 dimensi (fitur)**. Walau tidak terlalu banyak, ada dua alasan kuat mengapa tetap pertimbangkan PCA:

1.   Visualisasi

        *   Data berdimensi tinggi sulit divisualisasikan
        *   PCA memungkinkan kita menampilkan data dalam 2D atau 3D tanpa kehilangan terlalu banyak informasi.

2.   Mengurangi Noise atau Redundansi

        *   PCA menggabungkan fitur-fitur yang memiliki korelasi tinggi.
        *   Bisa membuat struktur klaster lebih jelas dan efisiensi untuk diproses.

dalam dataset iris, kombinasi `petal_legth` dan `petal_width` sudah sangat kuat untuk membedakan kelas, jadi PCA dapat membantu menonjolkan informasi utama.

##### **PCA dengan 2 dimensi**

````python
# PCA dengan 2 komponen
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_scaled)

# Konversi ke DataFrame
df_pca_2 = pd.DataFrame(X_pca_2, columns=['PCA1', 'PCA2'])

print("\nData setelah PCA (2 Komponen):")
print(df_pca_2.head())
````

##### **PCA dengan 3 dimensi**

````python
# PCA dengan 3 komponen
pca_3 = PCA(n_components=3)
X_pca_3 = pca_3.fit_transform(X_scaled)

# Konversi ke DataFrame
df_pca_3 = pd.DataFrame(X_pca_3, columns=['PCA1', 'PCA2', 'PCA3'])

print("\nData setelah PCA (3 Komponen):")
print(df_pca_3.head())
````

##### **Mengapa hasil PCA Bisa Bernilai Negatif?**

PCA **tidak mempertahankan nilai asli** dari fitur, melainkan membentuk **kombinasi linier baru** dari semua fitur yang sudah dinormalisasi. Nilai-nilai tersebut bisa:

*   **Positif**: jika kombinasi fitur mengarah ke satu sisi dari sumbu baru.
*   **Negatif**: jika mengarah ke sisi berlawanan dari sumbu tersebut.

**Ibarat Kompas atau Peta Baru**

Bayangkan jika punya peta yang biasanya arah utara ke atas. Sekarang putarlah peta itu sedikit (karena PCA mencari arah penyebaran terbaik). Titik-titik yang sebelumnya ke kanan sekarang bisa jadi berada di kiri peta baru, itu artinya nilainya jadi negatif di arah baru (PCA).

**Contoh Kasus Pada Iris Data**

Misalnya, PCA1 adalah gabungan dari `petal_length` dan `petal_width`:

*   Kalau suatu bunga punya ukuran yang **kecil**, maka nilainya di PCA1 bisa negatif karena jauh dari rata-rata.
*   Bunga lain yang punya ukuran besar akan punya nilai positif.




### **3. Modeling**

#### **K-Means Clustering**

Untuk proses modeling, saya menggunakan k-means clustering dengan jumlah cluster 2, 3, dan 4 untuk menghitung nilai evaluasi seperti **inertia** dan **silhouette score**, yang bisa membantu membandingkan kualitas clustering.

##### **K-Means dengan 2 Cluster**

````python
# KMeans Clustering dengan 2 cluster
kmeans_2 = KMeans(n_clusters=2, random_state=42, max_iter=700)
labels_2 = kmeans_2.fit_predict(df_scaled)

# Menampilkan centroid untuk 2 cluster
print("Centroid untuk 2 Cluster:")
print(kmeans_2.cluster_centers_)

# Menampilkan jumlah data per cluster untuk 2 cluster
print("\nJumlah data per cluster (2 Cluster):")
print(pd.Series(labels_2).value_counts())

# Visualisasi 2D
plt.figure(figsize=(6, 5))
plt.scatter(df_pca_2['PCA1'], df_pca_2['PCA2'], c=labels_2, cmap='Set1', s=50)
plt.title("K-Means (2 Cluster) - PCA 2D")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.grid(True)
plt.show()

# Visualisasi 3D
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_pca_3['PCA1'], df_pca_3['PCA2'], df_pca_3['PCA3'], c=labels_2, cmap='Set1', s=50)
ax.set_title("K-Means (2 Cluster) - PCA 3D")
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
plt.show()
````

##### **K-Means dengan 3 Cluster**

````python
from scipy.spatial.distance import cdist

# KMeans Clustering dengan 3 cluster
kmeans_3 = KMeans(n_clusters=3, random_state=42, max_iter=700)
labels_3 = kmeans_3.fit_predict(df_scaled)

# Hitung jarak dari tiap data ke setiap centroid
distances = cdist(df_scaled, kmeans_3.cluster_centers_)

# Buat DataFrame hasil
df_result = df_combined[['id']].copy()  # Ambil ID dari data asli
df_result['cluster'] = labels_3
df_result['dist_to_cluster_0'] = distances[:, 0]
df_result['dist_to_cluster_1'] = distances[:, 1]
df_result['dist_to_cluster_2'] = distances[:, 2]

# Tampilkan semua hasil
print("\n=== Hasil Perhitungan Lengkap (Semua Data) ===")
print(df_result.to_string(index=False))

# Menampilkan jumlah data per cluster untuk 3 cluster
print("\nJumlah data per cluster (3 Cluster):")
print(pd.Series(labels_3).value_counts())

# Visualisasi 2D
plt.figure(figsize=(6, 5))
plt.scatter(df_pca_2['PCA1'], df_pca_2['PCA2'], c=labels_3, cmap='Set2', s=50)
plt.title("K-Means (3 Cluster) - PCA 2D")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.grid(True)
plt.show()

# Visualisasi 3D
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_pca_3['PCA1'], df_pca_3['PCA2'], df_pca_3['PCA3'], c=labels_3, cmap='Set2', s=50)
ax.set_title("K-Means (3 Cluster) - PCA 3D")
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
plt.show()
````

##### **K-Means dengan 4 Cluster**

````python
# KMeans Clustering dengan 4 cluster
kmeans_4 = KMeans(n_clusters=4, random_state=42, max_iter=700)
labels_4 = kmeans_4.fit_predict(df_scaled)

# Menampilkan centroid untuk 4 cluster
print("Centroid untuk 4 Cluster:")
print(kmeans_4.cluster_centers_)

# Menampilkan jumlah data per cluster untuk 4 cluster
print("\nJumlah data per cluster (4 Cluster):")
print(pd.Series(labels_4).value_counts())

# Visualisasi 2D
plt.figure(figsize=(6, 5))
plt.scatter(df_pca_2['PCA1'], df_pca_2['PCA2'], c=labels_4, cmap='Set3', s=50)
plt.title("K-Means (4 Cluster) - PCA 2D")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.grid(True)
plt.show()

# Visualisasi 3D
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_pca_3['PCA1'], df_pca_3['PCA2'], df_pca_3['PCA3'], c=labels_4, cmap='Set3', s=50)
ax.set_title("K-Means (4 Cluster) - PCA 3D")
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")
plt.show()
````

### **4. Evaluasi**

#### **Evaluasi untuk Cluster 2, 3, dan 4**

````python
# Silhouette Scores
silhouette_2 = silhouette_score(df_scaled, labels_2)
silhouette_3 = silhouette_score(df_scaled, labels_3)
silhouette_4 = silhouette_score(df_scaled, labels_4)

# Inertia
inertia_2 = kmeans_2.inertia_
inertia_3 = kmeans_3.inertia_
inertia_4 = kmeans_4.inertia_

# Cetak hasil evaluasi
print("=== Evaluasi KMeans Clustering ===")
print(f"Cluster 2 - Silhouette Score: {silhouette_2:.4f} | Inertia: {inertia_2:.2f}")
print(f"Cluster 3 - Silhouette Score: {silhouette_3:.4f} | Inertia: {inertia_3:.2f}")
print(f"Cluster 4 - Silhouette Score: {silhouette_4:.4f} | Inertia: {inertia_4:.2f}")
````

#### **Perbandingan dengan Data Asli**

````python
# Buat label asli berdasarkan ID
def get_actual_label(idx):
    if idx < 50:
        return 'setosa'
    elif idx < 100:
        return 'versicolor'
    else:
        return 'virginica'

df_result['actual_label'] = df_result.index.map(get_actual_label)

# Tambahkan label hasil clustering
df_result['cluster'] = labels_3

# Buat mapping dari cluster ke label asli berdasarkan mode
label_mapping = {}
for cluster_id in range(3):
    mask = df_result['cluster'] == cluster_id
    majority_label = df_result.loc[mask, 'actual_label'].mode()[0]
    label_mapping[cluster_id] = majority_label

# Prediksi nama label berdasarkan hasil cluster
df_result['predicted_label'] = df_result['cluster'].map(label_mapping)

# Hitung jumlah prediksi yang benar
df_result['correct'] = df_result['actual_label'] == df_result['predicted_label']
jumlah_benar = df_result['correct'].sum()
total_data = len(df_result)
akurasi = jumlah_benar / total_data

# Tampilkan hasil evaluasi
print("\n=== Evaluasi Hasil Clustering vs Label Asli ===")
print(f"Total data    : {total_data}")
print(f"Prediksi benar: {jumlah_benar}")
print(f"Akurasi       : {akurasi:.4f}")
print("\nMapping cluster ke label:")
print(label_mapping)

# Tampilkan semua data yang salah prediksi
print("\n=== Semua Data yang Salah Prediksi ===")
print(df_result[~df_result['correct']].to_string(index=False))
````
